## Visualizing the Distribution of Continuous variables

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
year_columns = [col for col in df.columns if col.isdigit()]
decades = [1960, 1970, 1980, 1990, 2000, 2010, 2020, 2023]

# Create a DataFrame to hold decade-wise data
decade_data = pd.DataFrame()

for decade in decades:
    year_column = str(decade)
    if year_column in df.columns:
        decade_data[year_column] = df[year_column]
```

### Plotting Histograms Decade-wise

```python
for decade in decade_data.columns:
    plt.figure(figsize=(10, 6))
    plt.hist(decade_data[decade].dropna(), bins=20, color='skyblue', edgecolor='black')
    plt.xlabel('Population Value')
    plt.ylabel('Frequency')
    plt.title(f'Distribution of Population Values for {decade}s')
    plt.show()
```

## Visualizing the Distribution of Categorical variables

```python
well_known_countries = ['United States', 'China', 'India', 'Brazil', 'Russia', 'Japan', 'Germany', 'United Kingdom', 'France', 'Canada']

decades = [1960, 1970, 1980, 1990, 2000, 2010, 2020, 2023]
```

### Plotting Bar Charts Decade-wise

```python
for decade in decades:
    plt.figure(figsize = (12, 8))
    year_column = str(decade)
    if year_column in df.columns:
        
        df_filtered = df[df['Country Name'].isin(well_known_countries)]
        sns.barplot(x = 'Country Name', y = year_column, data = df_filtered, palette = 'pastel', hue = None)
        plt.xlabel('Country')
        plt.ylabel('Population')
        plt.title(f'Population Distribution by Well-Known Countries in {decade}')
        plt.xticks(rotation = 45, ha = 'right')  
        plt.legend().remove()  
        plt.show()
    else:
        print(f'No data available for the year {decade}')
```
